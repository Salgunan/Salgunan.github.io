```python
import pandas as pd
```

```python
df = pd.read_excel('Cleaned_Perovskite_Data2.xlsx')
print("The data is read")
```

```python
df = df.drop(["JV_forward_scan_PCE","JV_reverse_scan_PCE"], axis = 1)
print("columns are dropped")
```

```python
df.dropna(subset = ["Stabilised_performance_PCE", "Perovskite_composition_long_form"], inplace = True)
print("rows are dropped")
```

```python
df.shape
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = "median")
df_num = df.drop("Perovskite_composition_long_form", axis = 1)
imputer.fit(df_num)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(df_num)
print("transformation done")
```

```python
df_tr = pd.DataFrame(X, columns = df_num.columns)
print("df_tr prepared")
```

```python
df_cat = df[["Perovskite_composition_long_form"]]

df_cat.head(10)
```

```python
df_tr.reset_index(drop = True, inplace = True)
df_cat.reset_index(drop = True, inplace = True)

df_combined = pd.concat([df_tr, df_cat], axis = 1)
print("concantenated")
```

```python
df_combined.head()
```

```python
df_combined.to_excel('stabilised_performance_PCE_dataset.xlsx', index = False)
print("the dataset is saved to excel sheet")
```
